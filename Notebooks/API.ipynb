```python
# Now, to enrich the data on the dataset and also to ease the work at the "fake" logistics business, I have found the below API which includes a link to all the courier companies globally.
# As the company ships its products worldwide and I am supposedly part of the logistics department, this will allow the business to find a courier by just looking in the table all the options.
```

```python
import requests

url = "https://order-tracking.p.rapidapi.com/carriers"

headers = {
    'content-type': "application/json",
    'x-rapidapi-host': "order-tracking.p.rapidapi.com",
    'x-rapidapi-key': f'{token}''
    }

response = requests.request("GET", url, headers=headers)

print(response.text)
```

```python
envios = response.json()
```
